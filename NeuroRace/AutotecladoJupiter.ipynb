```python
import socket
import keyboard
UDP_IP = "192.168.0.101"  # Replace with ESP8266 IP
UDP_PORT = 4210

from pynput import keyboard
import time

# Flag to track whether the space key is currently pressed
space_pressed = False

def on_press(key):
    global space_pressed
    if key == keyboard.Key.space:
        space_pressed = True

def on_release(key):
    global space_pressed
    if key == keyboard.Key.space:
        space_pressed = False

# Start the listener in the background
listener = keyboard.Listener(on_press=on_press, on_release=on_release)
listener.start()

# Main loop checking key state
try:
    while True:
        if space_pressed:
            print("touche appuyée")
            MESSAGE = "1"
            sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
            sock.sendto(MESSAGE.encode(), (UDP_IP,UDP_PORT))
        else:
            print("aucune touche")
            MESSAGE = "0"
            sock = socket.socket(socket.AF_INET, socket.SOCK_DGRAM)
            sock.sendto(MESSAGE.encode(), (UDP_IP,UDP_PORT))
        time.sleep(0.2)  # Delay to avoid spamming
except KeyboardInterrupt:
    print("\nArrêt du programme.")
    listener.stop()
```
